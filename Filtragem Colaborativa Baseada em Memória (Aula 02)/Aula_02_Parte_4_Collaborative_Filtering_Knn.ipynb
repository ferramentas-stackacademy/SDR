# Usando Knn de Fato

```python
from sklearn.neighbors import NearestNeighbors
```

```python
data = pd.merge(movies, ratings, on = "movieId")
```

```python
data = data[["movieId", "title", "userId", "rating"]]
```

```python
# Define the features and target
X = data[['userId', 'movieId']]
y = data['rating']
```

```python
# Cria uma instância para NearestNeighbors 
#"kd_tree"
knn = NearestNeighbors(n_neighbors=10, algorithm="kd_tree")
```

```python
# Fit no modelo
knn.fit(X, y)

# Define o usuario e o filme no qual queremos fazer recomendações
user = 5
movie = 10
```

```python
movies[movies["movieId"]==10]
```

```python
# Achas os 10 vizinhos mais próximos para cada par de filmes e usuários 
distances, indices = knn.kneighbors([[user, movie]])
```

```python
# Pega as classificações desses usuários mais próximos
neighbor_ratings = y[indices[0]]
```

```python
# Calcula a média das avaliações que eles deram
average_rating = sum(neighbor_ratings) / len(neighbor_ratings)
```

```python
# Mostra a recomendação
print(f'A recomendação pro usuário {user} em relação ao filme {movie} é {average_rating}')
```
